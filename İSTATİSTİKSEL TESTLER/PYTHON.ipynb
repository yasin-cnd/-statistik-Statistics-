# UYUM İYİLİĞİ TESTLERİ


Uyum iyiliği İstatistiksel modelin, gözlem setine ne kadar iyi uyulduğunu açıklar. Uygunluk iyiliğinin ölçütleri genel olarak gözlemlenen değerler ile söz konusu modelde beklenen arasındaki tutarsızlığı özetmelemektedir. Bu ölçütler istatistiksel hipotez testi işleminde örneğin; hatalı modellerin normalleştirilmesinin testinde, özdeşleşmiş dağılımlardan çıkarılan iki örneklemin aynı olup olmadığının testinde (Bkz. Kolmogorov-Smirnov sınaması), sonuç frekanslarının belirli bir dağılımı takip edip etmediğinin testinde (Bkz. Pearson ki-kare testi) kullanılabilir. Varyans analizinde, varyansın içerisindeki değişkenlerden biri karelerinin toplamının bir bölümünü oluşturabilir.

#  Kolmogorov-Smirnov 

Kolmogorov-Smirnov testi, iki olasılık dağılımının (veya bir örnek dağılımı ve teorik bir dağılımın) aynı dağılımdan gelip gelmediğini değerlendirmek için kullanılan bir istatistiksel testtir. Bu test, bir veri setinin, belirli bir teorik dağılıma (genellikle normal dağılım) uygun olup olmadığını kontrol etmek için kullanılır.50'den büyük örneklemlerde daha iyi sonuçlar vermektedir.

PYTHON

```python
import numpy as np
import pandas as pd
import seaborn as sns
df = pd.read_csv("advertising (1).csv")
```

```python
df["Sales"]
```

SORU : Yukarıda verilen değişkenin dağılımının normal dağılıma uygunluğunu %95 güven düzeyinde araştırınız.


Ho: Örneklemin çekildiği kitlenin dağılımı ile normal dağılım arasında fark yoktur.


Hs: Örneklemin çekildiği kitlenin dağılımı ile normal dağılım arasında fark vardır.

```python
from scipy.stats import kstest

statistic, p_value = kstest(rvs=df["Sales"], cdf="norm", alternative='two-sided', mode='auto')

print("p_value :",p_value)
```

# Shapiro–Wilk testi

Shapiro-Wilk testi, bir veri setinin normal dağılıma uygun olup olmadığını değerlendirmek için kullanılan bir istatistiksel testtir. Bu test, veri setinin normal dağılımdan sapma derecesini ölçer ve normal dağılıma uyumunu değerlendirmek için bir istatistiksel ölçüt sağlar.Shapiro-Wilk testi, veri setinin normal dağılıma uygun olup olmadığını belirlemek için verilerin sıralı şekilde değerlendirilmesini kullanır. Test istatistiği, veri setinin normal dağılıma uyumunu temsil eder. Test istatistiği, normal dağılıma daha yakın olduğunda Shapiro-Wilk testi sonucu daha yüksek bir p değeri verir ve normal dağılıma daha uzak olduğunda p değeri düşer.

# Anderson-Darling Testi

Anderson-Darling testi, bir veri setinin normal dağılıma ne kadar uyduğunu değerlendirmek için kullanılan bir istatistik testidir.

Ho : Veri seti, normal bir dağılıma sahiptir.
    
H1 : Veri seti, normal bir dağılıma sahip değildir.

```python
from scipy.stats import anderson

data = [77, 91, 81, 84, 88, 69, 68, 76, 87, 78, 85, 82, 80, 83, 90]

result = anderson(data)
print("Anderson-Darling İstatistiği:", result.statistic)
print("Kritik Değerler:", result.critical_values)
print("P-değeri:", result.significance_level)
```

test istatistiği < kritik değer olduğundan ho reddedilemez.Veri seti, normal bir dağılıma sahiptir.

.

#  Ki-Kare Testi

Ki kare testi, istatistiksel analizlerde kullanılan bir yöntemdir ve iki kategorik değişken arasında ilişkinin olup olmadığını değerlendirmek için kullanılır. Bu test, gözlemlenen frekanslar ile beklenen frekanslar arasındaki farkı değerlendirerek istatistiksel bir sonuç üretir.

Ki kare testi, adını Pearson'un ki kare dağılımından alır. Ki kare dağılımı, pozitif olmayan değerler alan ve serbestlik derecesine bağlı olarak şekil değiştiren bir olasılık dağılımıdır. Ki kare testi, gözlemlenen ve beklenen frekanslar arasındaki farkı ki kare dağılımıyla karşılaştırarak test istatistiğini hesaplar.

Ki kare testi, genellikle "gözlemlenen" ve "beklenen" frekans tablolarıyla çalışır. Gözlemlenen frekanslar, gerçek veri setinde gözlemlenen kategorik değişken değerleridir. Beklenen frekanslar ise, iki değişken arasındaki bağımsızlık durumunda, her kategorinin beklenen frekansının ne olacağını tahmin etmek için kullanılan bir modelden elde edilir.

Hs: Eşlerin eğitim durumları birbiriyle bağımlıdır.(Kadının eğitim düzeyi ile erkeğin eğitim düzeyi arasında ilişki vardır.)

#  ANOVA


ANOVA (Analysis of Variance), istatistiksel bir yöntemdir ve birden fazla grup arasındaki farklılıkları incelemek için kullanılır. ANOVA, gruplar arasındaki ortalamalar arasında istatistiksel olarak anlamlı bir fark olup olmadığını belirlemek için kullanılır.
ANOVA, gruplar arasındaki farklılıkları, grup içi değişkenliğe kıyasla gruplar arası değişkenliği dikkate alarak analiz eder. Bu yöntemle, gruplar arasındaki varyansın grup içi varyansla karşılaştırılması sağlanır. Eğer gruplar arasındaki farklar, beklenenden daha büyük ise, gruplar arasında istatistiksel olarak anlamlı bir fark olduğu söylenebilir.

H0: Tüm grupların popülasyon ortalamaları arasında istatistiksel olarak anlamlı bir fark yoktur.
    
H1: En az bir grup, diğer gruplardan farklı bir popülasyon ortalamasına sahiptir.

```python
from scipy.stats import f_oneway

data = {'A': [88, 80, 79, 88, 76, 67, 98, 72, 77], 
        'B': [69, 71, 76, 67, 78, 89, 79, 72, 67], 
        'C': [58, 69, 64, 60, 69, 80, 81, 75, 70]}

statistic, p_value = f_oneway(data["A"], data["B"], data["C"])

print("ANOVA Test İstatistiği:", statistic)
print("p değeri:", p_value)
```

p<0.05 olduğundan Ho reddedilir. En az bir grup, diğer gruplardan farklı bir popülasyon ortalamasına sahiptir.

.

# LSD (Least Significant Difference) Testi

LSD (Least Significant Difference) Testi, ANOVA (Analysis of Variance) sonrasında çoklu karşılaştırmalar yapmak için kullanılan bir istatistiksel test yöntemidir. Bu test, ANOVA'da elde edilen grup ortalamaları arasındaki farkların istatistiksel olarak anlamlı olup olmadığını değerlendirmek için kullanılır. LSD testi, gruplar arasında tüm mümkün çift kombinasyonlarının farklarını karşılaştırır ve grup ortalamalarının arasındaki en küçük anlamlı farkı belirlemeye çalışır.

Ho: Grup ortalamaları arasında anlamlı bir fark yoktur.

H1: En az bir grup, diğer gruplardan istatistiksel olarak farklıdır.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.multicomp import MultiComparison

data =  {'A': [88, 80, 79, 88, 76, 67, 98, 72, 77], 
        'B': [69, 71, 76, 67, 78, 89, 79, 72, 67], 
        'C': [58, 69, 64, 60, 69, 80, 81, 75, 70]}
df = pd.DataFrame(data)

df_melt = pd.melt(df.reset_index(), id_vars=['index'], value_vars=['B', 'A', 'C'])
df_melt.columns = ['index', 'Treatment', 'Value']

mc = MultiComparison(df_melt['Value'], df_melt['Treatment'])
result = mc.tukeyhsd()
print(result)
```

LSD testi altında yer alan ikili karşılaştırma sonuçlarına göre

A ile B arasında (p=0.2547) (sig>α ise H0 KABUL) fark yoktur.

A ile C arasında (p=0.0246) (sig<α ise H0 RED) fark vardır.

B ile C arasında (p=0.466) (sig>α ise H0 KABUL) fark yoktur.

.

# Welch Testi

Welch ANOVA, varyansların eşit olmadığı durumlarda kullanılan bir varyans analizi yöntemidir. Bu test, gruplar arasındaki farklılıkları değerlendirirken varyansların eşit olmadığını varsayar ve buna göre istatistiksel analiz yapar.

Ho: Grupların popülasyon ortalamaları arasında istatistiksel olarak anlamlı bir fark yoktur.

H1: Grupların popülasyon ortalamaları arasında istatistiksel olarak anlamlı bir fark vardır.

```python
import pandas as pd
import numpy as np
import pingouin as pg

df = pd.DataFrame({'score': [64, 66, 68, 75, 78, 94, 98, 79, 71, 80,
                             91, 92, 93, 90, 97, 94, 82, 88, 95, 96,
                             79, 78, 88, 94, 92, 85, 83, 85, 82, 81],
                   'group': np.repeat(['strat1', 'strat2', 'strat3'],
                                      repeats=10)})
  
pg.welch_anova(data=df, dv='score', between='group')
```

p<0.05 olduğundan Ho reddedilir.Grupların popülasyon ortalamaları arasında istatistiksel olarak anlamlı bir fark vardır.

.

# Mann-Whitney U Testi

Mann-Whitney U testi, iki bağımsız grup arasında sıralı veya sıralanabilir verilerin dağılımı arasında istatistiksel olarak anlamlı bir fark olup olmadığını değerlendirmek için kullanılan bir parametrik olmayan testtir. 

Ho : Gruplar arasında fark yoktur.

H1 : Gruplar arasında fark vardır.

```python
import pandas as pd
from scipy.stats import mannwhitneyu

data = {'D': [22,25,26,24,19,20], 
        'E': [15,20,18,21,23,25]}
df = pd.DataFrame(data)

group1 = df['D']
group2 = df['E']

statistic, p_value = mannwhitneyu(group1, group2, alternative='two-sided')
print("Test İstatistiği:", statistic)
print("p-değeri:", p_value)
```

p>0.05 olduğundan Ho reddedilemez. Gruplar arasında fark yoktur.

.

# Kruskal-Wallis Testi

İstatistik bilim dalında Kruskal-Wallis sıralamalı tek yönlü varyans analizi, bağımsız gruplar arası anakütle medyanlarının eşitliğini sınamak amacı ile kullanılan bir parametrik olmayan istatistik sınamasıdır. Adı bu yöntemi ilk defa ortaya koyan William Kruskal ve W. Allen Wallis atıfla konmuştur. Matematiksel olarak ayrı olmakla beraber, tek yönlü varyans analizinin bir değişik şekli olarak görülebilir. Diğer bir görüşe göre Mann-Whitney U sınamasının 3 veya daha çoklu gruplara genişletilmesidir.

Ho: Gruplar arasında anlamlı bir fark yoktur.

H1: En az bir grup, diğer gruplardan istatistiksel olarak farklıdır.

```python
import pandas as pd
from scipy.stats import kruskal

data = {'F': [77, 81, 88, 74, 78, 79, 65, 58, 88, 88, 85, 86, 85, 86, 80],
        'G': [72, 66, 70, 82, 78, 69, 81, 86, 75, 74, 62, 79, 81, 89, 88],
        'H': [87, 96, 76, 84, 91, 99, 68, 79, 80, 79, 69, 60, 74, 76, 84],
        'I': [55, 56, 57, 58, 59, 60, 61, 62, 63, 64, 65, 62, 63, 68, 69]}

df = pd.DataFrame(data)

# Kruskal-Wallis testini gerçekleştirme
statistic, p_value = kruskal(df['F'], df['G'], df['H'],df["I"])

# Sonuçları yazdırma
print("Test İstatistiği:", statistic)
print("p değeri:", p_value)
```

p<0.05 olduğundan Ho reddedilir.En az bir grup, diğer gruplardan istatistiksel olarak farklıdır.

.

.

p<a olduğundan Ho reddedilir.En az bir grup, diğer gruplardan farklı bir popülasyon ortalamasına sahiptir.farklılığı yaratan kısmı bulmak için çoklu karşılaştırma testlerine bakalım.

LSD

Ho: Grup ortalamaları arasında anlamlı bir fark yoktur.

H1: En az bir grup, diğer gruplardan istatistiksel olarak farklıdır.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.multicomp import MultiComparison

data = {'okuyarak': [77, 91, 81, 84, 88, 69, 68, 76, 87, 78, 85, 82, 80, 83, 90], 
        'dinleyerek': [72, 76, 80, 82, 74, 67, 87, 81, 79, 79, 68, 77, 83, 88, 84], 
        'yazarak': [57, 69, 66, 74, 81, 92, 78, 79, 80, 69, 59, 60, 74, 76, 84]}
df = pd.DataFrame(data)

df_melt = pd.melt(df.reset_index(), id_vars=['index'], value_vars=['okuyarak', 'dinleyerek', 'yazarak'])
df_melt.columns = ['index', 'Treatment', 'Value']

mc = MultiComparison(df_melt['Value'], df_melt['Treatment'])
result = mc.tukeyhsd()
print(result)
```

okuyarak öğrenme metodu ile yazarak öğrenme metodu arasında (p=0.0195) (sig<α ise H0 RED) fark vardır.

Gruplar Arası Fark Kontrolü: 

Ho: Teknik yazarlık uzmanlık seviyesi bakımından, disiplinler arasında fark yoktur. (Gruplar arasındaki fark önemsizdir.)

Hs: Teknik yazarlık uzmanlık seviyesi bakımından, disiplinler arasında fark vardır.(Gruplar arasındaki fark önemlidir.)

```python
from scipy.stats import chi2_contingency

table = [[10,13,35],[47,12,9],[11,17,25]]
chi2, p, dof, expected = chi2_contingency(table)
a=0.05

print("Ki kare istatistiği:", chi2)
print("P değeri:", p)
print("Serbestlik derecesi:", dof)
print("Beklenen frekanslar:")
for row in expected:
    print(row)
    
if p > a:
    print('Ho reddedilemz')
else:
    print('Ho reddedilir')
```

p=0,000 <  0,05 olduğu için gruplar arası fark hipotezi reddedilir.

Teknik yazarlık uzmanlık seviyesi bakımından, disiplinler arasında fark vardır (Gruplar arasındaki fark önemlidir) denilebilir ya da teknik uzmanlık seviyesi bakımından disiplinler arasında fark olduğu %95 güvenle söylenebilir.
Farklılığı yaratan sınıfları tespit etmek için sınıflara ait ki-kare değerlerine ihtiyaç vardır.
Bu kısımın el ile hesaplanması gerekmektedir. Her satır için kendi içinde ki-kare istatistiği tekrar hesaplanır.

X1_2=4.11+1.70+1.03=7.43

X2_2=17.4+1+11.29=29.69

X3_2=6.54+0.01+7.08=13.63


Toplamları test istatistiği değerine (50.217) yaklaşık olarak eşittir.

Ki-kare değerlerine bakarak en farklı (en büyük) grubun 2.disiplin olduğunu söylemek mümkündür.
En büyük farklılığı yaratan grup olan 2. disiplin analiz dışı bırakılarak, kalan 2 disiplin arasında fark olup olmadığını araştırmak için analize devam edilir. Çıkarılmasına karar verilen disiplin, veri girişi aşamasında çıkarılmalı ve bütün işlemler tekrarlanmalıdır.

Bağımsızlık Kontrolü:

Ho: Teknik yazarlık uzmanlık seviyesi ile disiplin arasında ilişki yoktur.
(Değişkenler arasında ilişki yoktur.)

Hs: Teknik yazarlık uzmanlık seviyesi ile disiplin arasında ilişki vardır.
	(Değişkenler arasında ilişki vardır.)


```python
from scipy.stats import chi2_contingency

table = [[10,13,35],[11,17,25]]
chi2, p, dof, expected = chi2_contingency(table)
a=0.05

print("Ki kare istatistiği:", chi2)
print("P değeri:", p)
print("Serbestlik derecesi:", dof)
print("Beklenen frekanslar:")
for row in expected:
    print(row)
    
if p > a:
    print('Ho reddedilemz')
else:
    print('Ho reddedilir')
```

Yukarıda da görüleceği üzere 5’ten küçük sıklık yoktur. 
Burada 0.363>0.05 olduğundan Ho reddedilemez ve değişkenler arasında ilişki yoktur denilebilir.

Bu sonucu bağlı olarak, ilişki anlamsız olduğu için, ilişki katsayısının yorumlanmasına gerek yoktur.Ancak,
Araştırmada satır ve sütundaki nitel değişkenler sıralanabilir (ordinal) yapıda olduğu için, uygun olan ilişki katsayılarından  kendall tau incelenebilir.


H0: İki değişken arasında anlamlı bir ilişki yoktur.

H1: İki değişken arasında anlamlı bir ilişki vardır.

```python
import numpy as np
from scipy.stats import kendalltau

x = np.array([10,13,35])
y = np.array([11,17,25])

tau_coef, p_value = kendalltau(x, y)

print("p değeri:", p_value)
```

Burada da görüleceği üzere p>a (0.33>0.05) olduğundan Ho reddedilemez.İki değişken arasında anlamlı bir ilişki yoktur.
