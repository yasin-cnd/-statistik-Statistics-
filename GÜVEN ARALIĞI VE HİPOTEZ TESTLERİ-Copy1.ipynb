#                                              GÜVEN ARALIKLARI 

Parametresi bilinmeyen bir kitleden çekilen örneklemden hesaplanan istatistikler yardımıyla parametreyi tahmin etmenin iki yolu vardır.  Bunlardan birincisi,parametrenin tek bir değer olarak tahmin edildiği "nokta tahmini" konusudur. Bilinmeyen bir Q parameresinin nokta tahmini, parametrenin gerçek değerine eşit değildir.Çok sayıda rasgele örneklem alsak bile bu tahminin gerçek parametre değerine eşit olması çok zordur. Nokta tahmini,bize tahminin ne kadar bilgiye dayandığını söylemediği gibi,olası yanılma payı konusunda da fikir vermez.Dolayısıyla,Q parametresi için tek bir değer bulmak yerine,belirli bir güven düzeyinde Q'nın olası değerlerine ilişkin bir aralık vermek daha uygun olacaktır.Bu da bizi parametre için "aralık tahmini" kavramına götürecektir.Aralık tahmini, bir parametrenin olası değerlerini belirli bir güven düzeyiyle tahmin etmeye yönelik bir istatistiksel yöntemdir. Güven aralığı ise aralık tahmini sonucunda elde edilen aralığın güven düzeyini ifade eden bir istatistiksel kavramdır.

Güven aralıkları, bir veri kümesi veya örneklem için istatistiksel olarak hesaplanan bir parametrenin (örneğin, ortalama veya varyans) gerçek değerinin olası bir aralığını belirlemek için kullanılan bir yöntemdir. Güven aralığı, bir parametrenin olası değerlerini belirlemede bize bir aralık sağlar ve bu aralığın belirli bir güven düzeyinde (%95 gibi) doğru olduğu kabul edilir.
Güven aralıkları aşağıdaki şekilde özetlenebilir:

- Güven aralığı, bir parametrenin olası değerlerinin bir aralığını belirler.
- Güven aralığı, belirli bir güven düzeyi ile belirlenir (%95 gibi).
- Güven aralığı, veri kümesindeki örneklemlerden elde edilen istatistiksel ölçümlere dayanır.
- Güven aralığı, parametre tahminlerinin doğruluğunu ve farklı örneklemler arasındaki farklılıkları değerlendirmek için kullanılır.
- Güven aralığı, istatistiksel hipotez testlerinde kullanılan varsayımların doğruluğunu kontrol etmek için kullanılır.
- Güven aralığı, istatistiksel analizlerde ve karar verme süreçlerinde önemli bir rol oynar.
- Güven aralığı, parametre tahminlerinin belirli bir aralıkta olduğunu ve bu aralığın belirli bir güven düzeyi ile doğru olduğunu belirler.
- Güven aralığı, istatistiksel analizlerde ve karar verme süreçlerinde kullanılır.

.

- BAZI KAVRAMLAR

1. Güven aralığı: Kitle parametresi için belli bir güven düzeyinde hesaplanan aralık tahminidir.

2. Güven düzeyi: 100 1 − 𝛼 % ile gösterilir.Aralık tahmininin kitle parametresini kapsama olasılığını gösterir. Genellikle %99, %95, %90 kabul edilir. Güven düzeyi, olası aralığın ne kadar doğru olduğuna dair bir ölçüdür. Örneğin, %95 güven düzeyi, olası aralığın beklenen değerin gerçek değerinin %95 olasılıkla içinde olduğunu ifade eder.

3. Hipotez: Doğruluğu henüz kanıtlanmamış iddialara (önermelere) hipotez denir.

4. Hipotez testi: Örneklem istatistiklerinden yararlanarak bir hipotezin geçerli olup olmadığını ortaya koyma işlemine istatistiksel hipotez testi veya hipotez testi denir.

5. Olasılık değeri (p değeri): Hipotez testinde, null hipotezin reddedilip reddedilmeyeceğini belirlemek için kullanılan bir ölçüdür. P değeri, bir test istatistiğinin null hipotez altında gözlemlenme olasılığını ifade eder. Eğer p değeri belirli bir alfa seviyesinden daha küçükse (genellikle 0.05), null hipotez reddedilir ve alternatif hipotez kabul edilir. Eğer p değeri alfa seviyesinden büyükse, null hipotez kabul edilir ve alternatif hipotez reddedilir.

6. α : I.Tip hata ya da anlamlılık düzeyi (significance level) olarak bilinir. Hipotez testlerinde karşılaşabileceğimiz bir risk,gerçekte H0 hipotezi doğru iken, hipotez test sonucunda H0 hipotezini doğru olmadığını bulmaktır. Buna hata, istatistikte I. Tip hata denir. Araştırmacılar bunu çalışmanın başında kontrol altında tutmak isterler bu nedenle uygulamada genelde α=0,01, 0,05 ya da 0,10 kabul edilir.

7. Test istatistiği: null hipotezin doğru olup olmadığını belirlemek için kullanılır

8. Kitle : Bir araştırma veya analiz yapılmak istenen tüm bireylerin veya birimlerin oluşturduğu topluluktur.

9. Örneklem : Bir kitleden rastgele seçilen ve analiz için kullanılan küçük bir alt kümedir.Örneklem, kitleyi temsil edebilecek şekilde seçilmelidir , bu şekilde kitle hakkında genellemeler yapmak mümkündür.

10. Tip I hata: Null hipotezin doğru olduğu halde yanlışlıkla reddedildiği hata türüdür. Alfa seviyesi ne kadar düşük olursa, tip I hata yapma olasılığı da o kadar düşüktür.

11. Tip II hata: Null hipotezin yanlış olduğu halde yanlışlıkla kabul edildiği hata türüdür. Tip II hata, beta seviyesi ile ölçülür ve alfa seviyesi ile ters orantılıdır.

12. Z tablosu : standart normal dağılımın (ortalaması 0 ve standart sapması 1 olan) kümülatif dağılım fonksiyonunun belirli Z değerleri için karşılık gelen olasılık veya yüzde değerlerini gösteren bir tablodur.

13. T tablosu : t dağılımının belirli t değerleri için karşılık gelen olasılık veya yüzde değerlerini gösteren bir tablodur. t dağılımı, örneklem büyüklüğü n küçük olduğunda ve popülasyon standart sapması σ bilinmediğinde kullanılır.

14. Ki kare tablosu : Ki kare dağılımının belirli ki kare değerleri için karşılık gelen kuyruk olasılıklarını veya yüzde değerlerini gösteren bir tablodur. Ki kare dağılımı, popülasyonun dağılımını veya standart sapmasını bilmediğimiz durumlarda kullanılır

15. F tablosu, F dağılımının belirli F değerleri için karşılık gelen kuyruk olasılıklarını veya yüzde değerlerini gösteren bir tablodur. F dağılımı, iki popülasyonun varyanslarını karşılaştırmak veya regresyon analizi gibi çeşitli istatistiksel analizlerde kullanılır.

16. - S = örneklem standart sapması
    - S^2 = örneklem varyansı
    
    - σ = kitle standart sapması
    - σ^2 = kitle varyansı
    - n = veri adedi
    - x̄ = örneklem ortalaması
    - sd = serbestlik derecesi

```python
data1=np.array([1,2,3,4,5])
```

```python
import math
import statistics
from scipy.stats import norm

def Kitle_varyansının_bilindiği_durum(data): 
    alfa = 0.05
    n = len(data)       
    x̄ = np.mean(data)   
    σ = statistics.stdev(data)    
    z_tablo_degeri=norm.ppf(q=1-alfa/2)  
    
    alt_sinir = x̄ - z_tablo_degeri * ( σ / math.sqrt(n))
    üst_sinir = x̄ + z_tablo_degeri * ( σ / math.sqrt(n))
    güven_aralıgı=[alt_sinir,üst_sinir]
    return  güven_aralıgı

Kitle_varyansının_bilindiği_durum(data1)
```

```python
data4=np.array([58,60,65,60,63,70,55,67,58,69])
```

```python
import scipy.stats as stats
def kitle_varyansı_için_güven_aralığı(data):
    
    alfa = 0.05           
    n = len(data)         
    S_2 = statistics.variance(data) 
    χ2_1 = stats.chi2.ppf(alfa / 2 , n-1 )  
    χ2_2 = stats.chi2.ppf(1 - alfa / 2, n-1 )

    alt_sinir = (n - 1) * S_2 / χ2_2
    üst_sinir = (n - 1) * S_2 / χ2_1 
    güven_aralıgı=[alt_sinir,üst_sinir]
    return  güven_aralıgı

kitle_varyansı_için_güven_aralığı(data4)
```

1.3 KİTLE ORANI İÇİN GÜVEN ARALIĞI

```python
from math import sqrt
from scipy.stats import norm

def oran_guven_araligi(orneklem_buyuklugu, orneklem_orani, guven_seviyesi):
    
    z = norm.ppf(1 - (1 - guven_seviyesi) / 2)
    hata_payi = z * sqrt(orneklem_orani*(1-orneklem_orani)/orneklem_buyuklugu)
    alt_sinir = orneklem_orani - hata_payi
    ust_sinir = orneklem_orani + hata_payi
    güven_aralıgı=[alt_sinir,ust_sinir]
    return  güven_aralıgı

oran_guven_araligi(500,0.35,0.95)
```

```python
def Kitle_varyanslarının_bilindiği_durum(data1,data2):
    alfa=0.05
    n1 = len(data1)      
    n2 = len(data2)       
    x̄_1 = np.mean(data1)   
    x̄_2 = np.mean(data2)  
    σ_1 = statistics.stdev(data1)    
    σ_2 = statistics.stdev(data2)   
    z_tablo_degeri=norm.ppf(q=1-alfa/2)  
    
    alt_sinir = (x̄_1-x̄_2) - z_tablo_degeri * math.sqrt(((σ_1**2)/n1)+((σ_2**2)/n2))
    üst_sinir = (x̄_1-x̄_2) + z_tablo_degeri * math.sqrt(((σ_1**2)/n1)+((σ_2**2)/n2))
    güven_aralıgı=[alt_sinir,üst_sinir]
    return  güven_aralıgı

Kitle_varyanslarının_bilindiği_durum(data1,data2)
```

```python
import scipy.stats
def Kitle_varyanslarının_bilinmediği_durum_2(data1,data2):
    alfa=0.05
    n1 = len(data1)      
    n2 = len(data2)       
    x̄_1 = np.mean(data1)   
    x̄_2 = np.mean(data2)   
    σ_1 = statistics.stdev(data1)     
    σ_2 = statistics.stdev(data2)    
    
    T_tablo_degeri=scipy.stats.t.ppf(q=1-alfa/2,df=n1+n2-2)
    
    sp=math.sqrt((((n1-1)*(σ_1**2)) + ((n2-1)*(σ_2**2))) / (n1+n2-2))
    
    alt_sınır=(x̄_1 -x̄_2)-(T_tablo_degeri*sp*math.sqrt((1/n1)+(1/n2)))
    üst_sınır=(x̄_1 -x̄_2)+(T_tablo_degeri*sp*math.sqrt((1/n1)+(1/n2)))
    
    güven_aralıgı=[alt_sınır,üst_sınır]
    return  güven_aralıgı

Kitle_varyanslarının_bilinmediği_durum_2(data2,data3)
```

```python
def  İki_kitle_varyans_oranı_için_güven_aralığı(data1,data2):
    alfa=0.05
    n1=len(data1)
    n2=len(data2)
    S1=statistics.stdev(data1)
    S2=statistics.stdev(data2)
    F_1=scipy.stats.f.ppf(q=1-alfa/2, dfn=n1-1, dfd=n2-1)
    F_2=scipy.stats.f.ppf(q=1-alfa/2, dfn=n2-1, dfd=n1-1)
    alt_sınır=((S1*S1)/(S2*S2)) * (1/F_1)
    üst_sınır=((S1*S1)/(S2*S2)) * (F_2)
    güven_aralıgı=[alt_sınır,üst_sınır]
    return  güven_aralıgı
İki_kitle_varyans_oranı_için_güven_aralığı(data2,data3)
```
